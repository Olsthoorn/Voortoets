```python
import matplotlib.pyplot as plt
from matplotlib import patches
import numpy as np
from scipy.special import k0 as K0, j0 as J0, exp1 as Wt, jn_zeros
from scipy.integrate import quad
from scipy.signal import lfilter, filtfilt
import pandas as pd
from itertools import cycle


def newfig(title, xlabel, ylabel, xlim=None, ylim=None, xscale=None, yscale=None, figsize=None):
    fig, ax = plt.subplots(1, figsize=figsize)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    if xlim: ax.set_xlim(xlim)
    if ylim: ax.set_ylim(ylim)
    if xscale is not None: ax.set_xscale(xscale)
    if yscale is not None: ax.set_yscale(yscale)
    ax.grid(True)
    return ax

def Wh(u, rho=0):
    """Return Hantush's well function values.
    
    Parameters
    ----------
    u = r^s / (4 kD t)
    rho = r / lambda,    lambda = sqrt(kD c)
    
    >>>Wh(0.004, 0.03)
    4.894104204671381    
    """
    def kernel(y, rho):
        """Return the function to be integrated."""
        return np.exp(-y - (rho / 2) ** 2 / y) / y
    def w(u, rho): # Integrate the argument
        return quad(kernel, u, np.inf, args=(rho,))[0]
    wh = np.frompyfunc(w, 2, 1) # Vectorize function w(u, rho) so we can use arrays as input.
    return np.asarray(wh(u, rho), dtype=float)

Wh(0.004, 0.03)

def Wb(tau, rho=0):
    """Return Hantush well function values using the Bruggeman (1999) form of Hantush's Well function.
    
    Parameters
    ----------
    tau = t/ (cS)
    rho = r / lambda, lambda= sqrt(kD c)
    
    >>>Wb(0.05625, 0.03)
    4.894104204671381    
    """
    def kernel(x, rho):
        """Return the function to be integrated."""
        return np.exp(-x - (rho / 2) ** 2 / x) / x
    def w(tau, rho): # Integrate the argument
        return quad(kernel, 0, tau, args=(rho,))[0]
    wh = np.frompyfunc(w, 2, 1) # Vectorize function w(u, rho) so we can use arrays as input.
    return np.asarray(wh(tau, rho), dtype=float)


def Wb1(tau, rho=0):
    """Return Hantush well function values using the Bruggeman (1999) form of Hantush's Well function.
    
    Parameters
    ----------
    tau = t/ (cS)
    rho = r / lambda, lambda= sqrt(kD c)
    
    >>>Wb(0.05625, 0.03)
    4.894104204671381    
    """
    u = rho ** 2  / (4 * tau)
    return Wh(u, rho)


#u = np.logspace(-4, 1, 51)
u, rho = 0.004, 0.03
print("Wh(u={:.4g}, rho={:.4g}) = {}".format(u, rho, Wh(u, rho)))
tau = rho ** 2 / (4 * u)
print("Wb(tau={:.4g}, rho={:.4g}) = {}".format(tau, rho, Wb(tau, rho)))
```

# De nalijlende velaging (cumulatieve effect)

Hoeveel blijft er na verloop van tijd nog over van een tijdeijke onttrekking in het verleden?

De verlaging door een onttrekking in een situatie zonder randvoorwaarden wordt goed beschreven door Thuis

$$ \frac Q {2 \pi kD} W(u), \,\,\,\, u =\frac{r^2 S }{4 kD t} $$

Er is in een dergelijke situaite na beÃ«ndiging van de onttrekking nog zeer lange tijd sprake van een resterende verlaging, die samen met andere niet permanente onttrekkingen toch tot een cumulatief effect zal leiden.

De resterende verlaging is dan gelijk aan
 $$ s(r, t) = \frac{Q}{4 \pi kD} \left[ W_t \left(\frac{r^2 S}{4 kD t}\right) - W_t \left( \frac{r^2 S}{4 kD (t - \Delta t)} \right)\right]$$

Voorbeeld:


```python
kD, S = 650, 0.002 # [m2/d]
Dt = 120 # [d] Duur van de onttrekking
r = 1000 # [m]
t = np.logspace(0, 4, 121) # [d]

putten = {
    0: {'Q': +650., 'tStart':   0.},
    1: {'Q': -650., 'tStart': 120.}
}

s = np.zeros_like(t)
for k in putten:
    put = putten[k]
    Q, tStart = put['Q'], put['tStart']
    u = r ** 2 * S  / (4 * kD * (t[t > tStart] - tStart))
    s[t > tStart] += Q / (4 * np.pi * kD) * Wt(u)

fig, axs = plt.subplots(2, 1, figsize = (10, 5), sharex=True)
for ax in axs:    
    ax.grid()
    ax.plot(t, s,  color='k', label='Verlaging [m]')
    ax.legend()
axs[0].set(title='Naijling verlaging volgens Theis (kD = {:.0f} m2/d, S = {:.3g})'.format(kD, S),
           ylabel='verlaging [m]',
            yscale='linear', xscale='linear')
axs[1].set_ylim(0, 0.01)
axs[1].set_xlabel('tijd [d]')
axs[1].set_xlim(0, 4000)
    


```

# Verruijt vs Dupuit en De Glee

De formule van Verrijt:
Deze gaat ervan uit dat de onttrekking precies gelijk is aan de neerslag binnen de straal waarop de verlaging nul is. Het is wel heel toevallig wanneer de ontrekking en gebiedsrand met vaste stijghoogte met elkaar overeenkomen. 
Man kan ook zeggen dat die de verlaging berekent binnen een circulair intrekgebied, het gebied waarbinnen al het water naar de put stroomt. Het intrek gebied wordt begrenst door de cirkel waarop er geen verhang is naar de put. Dit is echter niet de grens van de verlaging die de put veroorzaakt.

$$ Q_r = - \left(h_0^2 - h_w^2 + \frac N {2k}(R^2 - r^2) \right)\frac{\pi k}{\ln \frac{r}{R}} $$

Eenvoudiger en direct in te zien: 

$$ Q_r = N \pi r^2 - Q$$

Met als oplossing van Verruijt voor onttrekking uit een pakket met vrije watertafel en vlakke bodem, uit het centrum van een gebied met neerslag :

$$ h(r)^2 = H^2 + \frac N {2k}(R^2 - r^2) + \frac{Q_0}{\pi k}\ln\frac{r}{R} $$

Het is duidelijk dat de radius waarop het verhang nul is, de grens van het intreggebied van de put wordt bepaald door

$$ Q = N \pi r^2 $$

Het zou wel heel toevallig zijn wanneer de onttrekking gelijk is aan de totale neerslag op het circulare gebied. Alleen dan valt het inttrekgebied van de put samen met de gebiedsrand waar $h=H$,is vast. Het plaatje bij de uitleg van de formule van Verrijt in de "Handleiding Berekeningsinstrument bemalingen van een bouwput", pag. 9. staat wel h(0)=H, de stijghoogte op de rand en er staat R is de invloedsstraal op vaste tot aan de put. 

Voorbeeld

```python
k, H, R, N, Q, rw = 10, 20, 1000, 0.001, 1200, 1.

def h(Q, r):
    return np.sqrt(H ** 2 + N / (2 * k) * (R ** 2 - r ** 2) + Q / (np.pi * k) * np.log(r  / R))

ax = newfig("Formule van Verruijt k={:.0f} m/d, H={:.0f} m, N={:.3g} m/d, R={:.0f} m".format(k, H, N, R), "r [m]", "h [m]",
            figsize=(10, 6))

clrs = cycle('brgmck')
Qs =[0.001, 300, 600, 1200, 1800, 2400]
for Q in Qs:
    clr = next(clrs)
    r = np.logspace(0, 3, 31)    
    ax.plot(-r, h(Q, r), color=clr, label='Q = {:.0f}'.format(Q))
    ax.plot(+r, h(Q, r), color=clr, label='')
    r = np.sqrt(Q / (np.pi * N))
    ax.plot(-r, h(Q, r), ls='none', marker='v', mfc=clr, mec='k', ms=6, label='intrekgrens', zorder=5)
    ax.plot(+r, h(Q, r), ls= 'none', marker='v', mfc=clr, mec='k', ms=6, label='', zorder=5)
ax.plot([+R, +R], [0, H], '--', color='blue', lw=2, label='vaste rand')
ax.plot([-R, -R], [0, H], '--', color='blue', lw=2, label='')
ax.plot([+rw, +rw], [0, H], ':', color='blue', lw=2, label='put')
ax.plot([-rw, -rw], [0, H], ':', color='blue', lw=2, label='')
ax.add_patch(patches.Rectangle((-R, -2), 2 * R, 2, fc='gray'))
leg = ax.legend(loc='lower left', fontsize=10)
leg.set_bbox_to_anchor((0.10, 0.11, 0.3, 0.5), transform=ax.transAxes)
```

In conclusie, de formule van Verruijt is evenzeer van toepassing als die van Dupuit. Waar Dupuit alleen de verlaging berekent of the natte dicte van de de freatische laag zonder neerslag, geeft de formule van Verruijt deze natte dikte van de freatische laag voor ontrekking en neerslag. De verlaging kan nu worden berekend door h voor Q=Q (met onttrekking) af te trekken van de h voor Q=0 (zonder onttrekking). Wat wellicht verwarring geeft als gevolg van de figuur in genoemde handleiding, is dat daar in de figuur de circle zonder verhang samenvalt met de cirkel met vaste rand. Dat is echter geenszins een beperking in de formule van Verruijt. Deze neemt binnen de vaste rand netjes neerslag en verlaging samen mee voor de situatie waarin de effectieve laagdikte significant verandert.

De verlating wordt bij Verruijt verkregen door de grondwaterstand met onttrekking af te trekken van die zonder onttrekking. Bij Dupuit en alle andere oplossingen met vaste laagdikte wordt direct de verlaging berekend. De relatie tussen de berekenigen met vaste laagdikte en die met freatische, ruimtelijke laagdikte volgt direct uit de formule van Verruijt door te schrijven:

$$ h^2 - H^2 = (h - H) (h + H) \approx (h - H) 2H = 2 H s $$

Dupuit met neerslag:
$$ s = \frac N {4kH}(R^2 - r^2) + \frac{Q_0}{2 \pi kH}\ln\frac{r}{R} $$

Verruijt
$$ s = \frac N {2k(h + H)}(R^2 - r^2) + \frac{Q_0}{\pi k (h + H)}\ln\frac{r}{R} $$

Bij Verruijt blijft de te berekenen $h$ nog in de noemer staan als je direct de verlaging wilt berekenen. Het verschil tussen beide zit dus in de afwijing van $h$ t.o.v. $H$. Bij een afwijking van 20% is de fout in de berekening ca. 10%; uiteraard kan direct Verruijt worden gebruikt of kan de verlaging iteratief worden berekend met de formule hierboven.

# Wanneer worden de verlagingen stationair?

Bij onttrekking met een put aan een oneidig uitgestrekt watervoerend paket ontstaat er nooit een stationaire situatie omdat er geen terugkoppeling is tussen verlagen en infiltratie.

Verlagingen kunnen alleen stationair worden wanneer ergens in het watervoernde pakket randen zijn met vaste stijghoogte, danwel er eenvaste stijghoogte is een een bovenliggend pakket waarmee via lek door een semi-doorlatende laag verbinding bestaat. De essentie is dat door de verlaging een een stroming wordt opgewekt die op termijn de onttrekking volledig kan compenseren.

De belangrijkste situatie waarbij stationaire stoming optreedt is meestal oppervlaktewater verderop in het gebied of een semi-permanente toplaag waarboven het oppervlaktewaterpeil ruimtelijk wordt beheerd.

Voor een punt die onttrekt uit een pakket met semi-gespannen water geldt de formule van Hantush (1955)

$$ s(r, t) = \frac{Q}{4 \pi kD} W_h \left(u, \rho\right), \,\,\,\,with\,\,\,\, u = \frac{r^2 S}{4 kD t},\,\,\,\,and\,\,\,\,\rho=\frac{r}{\lambda}$$

Deze verlaging heeft wanneer uitgezet op logartmische tijdschaal een buigpunt halverweg de eindverlaging, waar de verlaging op de logarithmische tijdas steeds trager gaat verlopen. Dit buigpunt is een geschikt moment om op praktische wijze de duur te bepalen waarop de verlaging stationair wordt. Het buigpunt ligt op

$$ u = \frac \rho 2 = \frac{r}{2 \lambda}$$

Om het snijpunt van de raaklijn door het buigpun en de eindverlaging te bepalen is het mathematisch handiger om direct met de de parameter op de horizontale as te werken

$$ \eta = \ln \left(\frac 1 u \right) = -\ln u $$

De tangent of the raaklijn door de het buigpunt op de as $1/u$ is

$$ \left(\frac{\partial W_h}{\partial \eta} \right) = e^{-\rho}$$

De stationaire eindwaarde is

$$W_h(0, \rho) = 2 K_0(\rho)$$


Waar de raaklijn door het buigpunt de lijn met de eindverlaging snijdt is de verlaging halverwege de eindverlaging behorend bij de stationaire situatie. Het punt waarop de verlaging stationair is geworden kan praktisch worden gedefinieerd als het punt waarop de raaklijn door het buigpunt de horizontale lijn met de eindverlaging snijdt.



$$ \eta = e^{\rho} K_0(\rho) + \eta_{Bp},\,\,\,\,with\,\,\,\,\eta = -\ln(u)$$

$$ \eta_{Bp} = -\ln \left(u_{Bp} \right),\,\,\,\, with\,\,\,\,u_Bp = \frac{\rho}{2}$$

$$\eta - \eta_Bp = e^{\rho} K_0(\rho)$$

$$\frac{e^\eta}{e^{\eta_{Bp}}} = e^{e^{\rho} K_0(\rho)}$$

$$\frac{u_{Bp}}{u} = e^{e^{\rho} K_0(\rho)}$$

$$ u = u_{Bp}  e^{-e^{\rho} K_0(\rho)}$$


In het voorbeeld hieronder is de Hantush functie getoond versus 1/u op logaritmische schaal. De dubbel buiging waarvan de eerste start wanneer de invloedsgrens van de onttrekking de waarnemingsput bereikt en de tweede wanneer de verlaging wordt afgeremd door de opgewekte lek. De eindverlaging is gelijk aan de stationaire verlaging, zodat, met $t\rightarrow\infty,\,\,u\rightarrow0$:

$$ W_h(0, \rho) = 2 K_0(\rho)$$

Het buigpunt light op halve hoogte, dus bij $W_h(u_{bp}) = K_0(\rho)$. Voor het buigpunt geldt bovendien

$$u_{bp} = \frac{\rho}{2} = \frac{r}{2 \lambda} $$

De berekende buigpunten worden met een dikke stip in de grafiek aangegeven.

Voorts is in de grafieken aangegeven de raaklijn aan de putfunctie door het buigpunt en waar deze de eindverlaging snijdt. Zowel deze raaklijn als de horizontale eindverlaging zijn in de grafiek opgenomen.

Het snijpunt van deze raaklijn met de eindverlaging ligt op

$$ u = u_{Bp}  e^{-e^{\rho} K_0(\rho)}$$

Deze gemakkelijk te berekenen waarde van $u$ is m.i. een praktische waarde voor het te definieren moment waarop de eindverlaging is bereikt.

$$ u = \frac{\rho}{2} e^{-{e^\rho} K_0(\rho)}$$

En voor een reÃ«le situatie, en mijn voorkeurs uittrukking omdat het argument in de Theis formule (geen lek) links staat en de invloed van de lek rechts:

$$u = \frac{r^2 S}{4 kD t}=\frac{r}{2 \lambda} e^{-{e^{\frac{r}{\lambda}} K_0\left(\frac{r}{\lambda}\right)}}$$

$$ u = \frac{r^2 S}{4 kD t}=\frac 1 2 \rho e^{-{e^{\rho} K_0(\rho)}}$$

```python
# Hantush op half log schaal
# Buigpunt, raaklijn en snijpunt met eindverlaging
# K0 erbij betrekken.

u = np.logspace(-5, 1, 51)
ax = newfig("Wantush's well function vs 1/u", "1/u", "Wh(u, rho)", xscale='log', figsize=(10, 5))

clrs = cycle('brgmck')
rhos = [0.01, 0.03, 0.1, 0.3, 1, 3]
for rho in rhos:
    clr = next(clrs)
    ax.plot(1/u, Wh(u, rho), color=clr, label=r'$\rho$= {:.2f}'.format(rho))
    uBp = rho  / 2    
    eta2 = np.exp(rho) * K0(rho) - np.log(uBp)
    u2 = np.exp(-eta2)
    ax.plot(1 / uBp, Wh(rho / 2, rho), 'o', mfc=clr,  label=r'buigpunt $\rho$={:.2f}'.format(rho), zorder=5.1)    
    ax.plot([1 / uBp, 1 / u2], [K0(rho), 2 * K0(rho)], '.--', color=clr, label='', zorder=4.9)
    ax.plot([1 / u2, 1 / u[0]], [2 * K0(rho), 2 * K0(rho)], '-.', color=clr, label='', zorder=5)
ax.legend(loc='upper left', fontsize=9)
```

Op logschaal betekent dat we met de variable $\eta = \ln{1/ u} = -\ln(u)$ werken. Dit gebeurt hieronder door dezelfde grafieken te genereren als hiervoor maar dan op lineaire schaal van $\eta = -\ln(u)$. De vergelijking van de lijn door het buigpunt (bp), dat lag op $u_{bp}=\frac \rho 2$, zodat $\eta = - \ln\left(\frac \rho 2\right)$

De vergelijking van de raaklijk in $\eta$ coordinaten:

$$ \eta = e^{\rho} K_0(\rho) + \eta_{Bp},\,\,\,\,with\,\,\,\,\eta = -\ln(u)$$

$$ \eta_{Bp} = -\ln \left(u_{Bp} \right),\,\,\,\, with\,\,\,\,u_Bp = \frac{\rho}{2}$$

Nogmaals de vergelijking

$$\eta - \eta_Bp = e^{\rho} K_0(\rho)$$

Neem aan beide zijden de exponent

$$\frac{e^\eta}{e^{\eta_{Bp}}} = e^{e^{\rho} K_0(\rho)}$$

Dan kunnen we terug naar de oorspronkelijke coordinaten in $u$

$$\frac{u_{Bp}}{u} = e^{e^{\rho} K_0(\rho)}$$

Of zo

$$ u = u_{Bp}  e^{-e^{\rho} K_0(\rho)}$$

Zodat het punt dat als bereiken stationariet kan worden beschouwd wordt verkregen door het de u=waarde van het buigpunt te vermenigvuldigen met de factor $e^{-e^{\rho} K_0(\rho)}$. Omdat $\rho > 0$ is deze factor altijd positief.

Zoals hiervoor reeds afgeleid hebben we voor het punt waarop stationariteit kan worden aanvaard het volgende:

$$ u = u_{Bp}  e^{-e^{\rho} K_0(\rho)}$$

Deze gemakkelijk te berekenen waarde van $u$ is m.i. een praktische waarde voor het te definieren moment waarop de eindverlaging is bereikt.

$$ u = \frac{\rho}{2} e^{-{e^\rho} K_0(\rho)}$$

En voor een reÃ«le situatie, en mijn voorkeurs uittrukking omdat het argument in de Theis formule (geen lek) links staat en de invloed van de lek rechts:

$$u = \frac{r^2 S}{4 kD t}=\frac{r}{2 \lambda} e^{-{e^{\frac{r}{\lambda}} K_0\left(\frac{r}{\lambda}\right)}}$$

# Hoe verloopt het debiet en de verlaging wanneer  op verlaging wordt bemalen?

Bij een constante onttrekking en weing randvoorwaarden blijft de verlaging toenemen. Bouwputten echter worden bemalen om een verlaging te behalen. ergo wordt na het behalen van de gewenste verlaging het onttrekkingsdebiet geleidelijk aan teruggeschroefd om de verlaging te behouden. Het onttrekkingsdebiet neemt dus af in de tijd, en dit heeft ook gevolgen voor de verlaging die in de omgeving optreedt.

Er bestaat echter geen analytische formule die uitgaat van een vaste verlaging in de put; er bestaan alleen oplossingen met een gegeven eventueel tijdsafhankelijk debiet. Maar wellicht is een van de oplossingen  een voldoende benadering van de situatie met vaste verlaging.

Bruggeman (1999) geeft divere formules voor de niet-stationaire verlaging in putten als functie van de tijd. Omdat het bemalingsdebiet na het bereiken van de vereiste drooglegging zal verminderen om de verlaging niet niet verder dan nodig te laten oplopen, lijkt de analytische oplossing voor een onttrekking met exponentieel afnemend debiet een geschikte kandiaat. Volgens Bruggeman (1999, oplossing 215.17, p160) geldt daarvoor als randvoorwaarde voor $r \downarrow 0$:

$$ Q(t) = Q_0 e^{-\alpha t}$$

met als oplossing

$$s (r, t) = \frac{Q_0 \, e^{-\frac t T}}{4 \pi kD} W\left(\frac t T,\, \sqrt \frac {r ^2 S} {kD T}\right) $$

De radius komt dan overeen met de representatieve straal van het te bemalen vlak.

Benadering van Theis


Situatie op $t=t_1$:

$$s_{t_1}=\frac{Q}{4\pi kD}\ln\left(\frac{2.25kD\left(t_{1}-t_{0}\right)}{r^{2}S}\right)$$

Situatie op $t_{2}$:

$$s_{t_2}=\frac{Q}{4\pi kD}\ln\left(\frac{2.25kD\left(t_{2}-t_{0}\right)}{r^{2}S}\right)-\frac{\Delta Q}{4\pi kD}\ln\left(\frac{2.25kD\left(t_{2}-t_{1}\right)}{r^{2}S}\right)$$

dus, $s_{t_1}=s_{t_2}$

$$\ln\left(\frac{2.25kD\left(t_{1}-t_{0}\right)}{r^{2}S}\right)=\ln\left(\frac{2.25kD\left(t_{2}-t_{0}\right)}{r^{2}S}\right)-\frac{\Delta Q}{Q}\ln\left(\frac{2.25kD\left(t_{2}-t_{1}\right)}{r^{2}S}\right)$$

$$\frac{\Delta Q}{Q}	=\frac{\ln\left(\frac{t_{2}-t_{0}}{t_{1}-t_{0}}\right)}{\ln\left(\frac{2.25kD\left(t_{2}-t_{1}\right)}{r^{2}S}\right)}$$

Hieronder toetsen we deze afleiding

```python
t0, t1, t2 = 0., 7., 97.
kD, S = 650, 0.2
s1, s2, r = 5., 5., 20.

Q = s1 * 4 * np.pi * kD / np.log(2.25 * kD * (t1 - t0) / (r ** 2 * S))
alpha = np.log((t2 - t0) /(t1 - t0)) / np.log(2.25 * kD * (t2 - t1) / (r ** 2 * S))

t = np.logspace(-1, 2, 301)
u1 = r ** 2 * S / (4 * kD * t)
s = Q / (4 * np.pi * kD) * Wt(u1)
ss = Q / (4 * np.pi * kD) * np.log(2.25 * kD * (t - t0) / (r ** 2 * S))
u2 = r ** 2 * S  / (4 * np.pi * kD * (t[t > t1] - t1))
s[t > t1] -= alpha * Q / (4 * np.pi * kD) * Wt(u2)
ss[t > t1] -= alpha * Q / (4 * np.pi * kD) * np.log(2.25 * kD *  (t[t > t1] - t1) / (r ** 2 * S))

ax = newfig('Bemaling', 'tijd [d]', 'verlaging [m]', xscale='log')
ax.plot(t, s, label='Theis')
ax.plot(t, ss, label='ln(2.25...)')

ax.legend()

plt.show()

```

Impulse response is de afgeleide van de stap response voor eenheidsonttrekking

De convolutie integraal vergt zeer kleine tijdstappen voor voldoende nauwkeurigheid. Daarom is ene betere aanpak om direct met de block impulse te werken.

$$ Br = \frac{1}{4 \pi kD} \left(W_t(\frac{r^2 S}{4 kD t}) - W_t(\frac{r^2 * S / {4 \pi kD (t[t > dt] - dt)}}) $$

# Vertraagde nalevering (Delayed yield)

Pompproeven die als basis voor aanvragen van vergunningen worden uitgevoerd, zijn vaak zo kort dat alleen de verlaging in het bepompte pakket kan worden bepaald, en geen informatie wordt verkregen van het later nazakken van het freatisch vlak boven de scheidende laag, die optreedt als gevolg van de door het pompten toegenomen lek. De voortoets zou altijd moeten eisen dat dit effect van nazakken wordt meegenomen in de analyse.

$$ s = \frac Q {4 \pi kD} W(u) = \frac Q {2 \pi kD} K_0 \left(\frac r \lambda \right) $$

$$ W(u) = 2 K_0 \left(\frac r \lambda \right) $$

$$ \ln \left( \frac {2.25 kD t}{r^2 S_y} \right) = 2 K_0 \left(\frac r \lambda \right) $$

$$t = \frac{r^s S_y} {2.25 kD} e^{2 K_0 \left( \frac r \lambda \right)}$$

De algemene situatie met vertraagde naverling kan voor het semi-gespannen pakket worden benaderd door de aanvankeijke verlaging volgens Hantush met de elastische bergingscoÃ«fficiÃ«nt S, gevolgd door de verlaging volgens Theis met de freatische bergingscoÃ«fficiÃ«nt wanneer de verlaging volgens Theis die volgens Hantush inhaalt. Vanaf dat moment zijn de verlaging in het ondiepe en diepe pakket geijk, en lopen samen verder op. In de werkelijkheid ontstaat een vloeiende overgang van de Hantush verlaging naar die volgens Theis, die met een tweelagen model kan worden berekend. Dat is wat ignewikkelder. Voor de praktijk kan de hier gevolgde methode als benadering worden gebruikt. Hij laat in elk geval zien hoe lang men een pompproef doen om zeker te weten dat er in later stadium geen substantiÃ«le daling van het freatisch vlak optreedt.


Voorbeeld:

```python
kD, Sy,S, c, Q0 = 650, 0.2, 0.002, 2500., 1200
L = np.sqrt(kD * c)

t = np.logspace(-3, 3, 61)
rs = [10, 30, 90]

ax = newfig("Delayed yield kD = {:.0f}, Sy = {:.3g}, S = {:.3g}, c = {:.0f} d".format(kD, Sy, S, c),
            "tijd [d]", "s [m]",xscale='log')
clrs = cycle('brgmck')
for r in rs:
    clr = next(clrs)
    rho = r  / L
    tx = r ** 2 * Sy / (2.25 * kD) * np.exp(2 * K0(rho))
    u1 = r ** 2 * Sy / (4 * kD * t)
    u2 = r ** 2 * S  / (4 * kD * t)    
    s1 = Q0 / (4 * np.pi * kD) * Wt(u1)
    s2 = Q0 / (4 * np.pi * kD) * Wh(u2, rho)    
    s2[t > tx] = s1[t > tx]
    ax.plot(t, s1, ls='--', color=clr, label='r = {:3.0f} m, S = {}, c = {:.0f} d'.format(r, Sy, c))
    ax.plot(t, s2, ls= '-', color=clr, label='r = {:3.0f} m, S = {}'.format(r, S))
ax.legend()
plt.show()

```

# Invloedszone (radius of influence)

Om de reikwijdte van een dyamische onttrekking aan te geven wordt wel gesproken van invloedsgrens of "radius of influence". De radius of influence bij een continue onttrekking in een zich oneindig uitstrekking pakket zonder lek kan direct worden afgeleid uit de verlaging volgens Theis, die in dit geval van toepassing is. De verlaging volgens Theis kan voor voldoende grote tijd worden benaderd met de vereenvoudigde formule

$$ s = \frac Q {4 \pi kD} \ln\left(\frac{2.25 kD t}{r^2 S}\right) $$

Voor het nog niet beÃ¯nvloede gebied geldt dat de verlaging nog nul is. Dit is hier het geval wanneer het argument onder de logaritme gelijk is aan 1:

$$ \frac{2.25 kD t}{r^2 S} = 1 $$

zodat aaar een radius of influence uit volgt gelijk aan

$$ r = \sqrt{\frac {2.25 kD t}{S}} $$ 

Deze formule is een eenvoudige doch uitermate praktisch middel om de radius te bepalen waarbuiten de invloed van de onttrekking nog niet bemerkbaar is. Het is een benadering, omdat de gebruikte formule van de verlaging een benadering is, maar het is een praktische benadering die op de aangegeven wijze objectief is afgeleid. Hieronder volgt een voorbeeld waarin de werking van deze radius inzichtelijk wordt gemaakt. 

```python
kD, S, Q0 = 650, 0.001, 2400

fig, axs = plt.subplots(2, 1, figsize=(10, 8), sharex=True )
axs[0].set_title('Theis and approximation')
axs[0].set_ylabel('s m')
axs[0].grid()
axs[0].set_xscale('log')
axs[0].set_ylim(7, -1)
axs[1].set_title('Raidus of influence')
axs[1].set_ylabel('tijd [d]')
axs[1].set_xlabel('r [m]')
axs[1].set_yscale('log')
axs[1].grid()

r = np.logspace(0, 5, 51)
ts = np.logspace(-2, 4, 5)

clrs = cycle('brgmck')
for t in ts:
    clr = next(clrs)
    u = r ** 2 * S  /( 4 * kD * t)
    sTheis = Q0 / (4 * np.pi * kD)* Wt(u)
    sLn    = Q0 / (4 * np.pi * kD) * np.log(2.25 * kD * t / (r ** 2 * S))
    axs[0].plot(r, sTheis, ls='-',  color=clr, label='Theis, t = {:.4g} d'.format(t))
    axs[0].plot(r[sLn > -0.5], sLn[sLn > -0.5],    ls='--', color=clr, label='Ln, t = {:.4g} d'.format(t))
    rinf = np.sqrt(2.25 * kD * t / S)
    axs[0].plot(rinf, 0, 'o', mfc=clr, mec=clr, label='r influence at t={:.4g} d'.format(t))
    axs[1].plot(rinf, t, 'o', mfc=clr, mec=clr, label='r influence at t={:.4g} d'.format(t))
    
arrow = patches.FancyArrowPatch((0.25, 0.65), (0.75,0.65), transform=axs[0].transAxes,
                                mutation_scale=100, fc='gray', ec='black', alpha=0.5)
#arrow = patches.FancyArrowPatch((0.25, 0.75), (0.75,0.75), transform=fig.transFigure, fc='gray', ec='black', alpha=0.5)
axs[0].add_patch(arrow)
axs[0].legend(loc='upper left', fontsize=8)

tt = np.logspace(-2, 4, 61)
rinf = np.sqrt(2.25 * kD * tt / S)
axs[1].plot(rinf, tt, label='Radius of influence')
axs[1].legend()
plt.show()
```

# Bemalingen (VariÃ«rende onttrekking om de verlaging te handhaven)

Bemalingen hebben als doel een verlaging te bereiken om droog te kunnen werken en bouwen. Er zal een voldoende hoog debiet worden ingezet om in een beperkte tijd de vereiste verlaging te halen. Daarna zal het debiet worden verminderd om de verlaging te behouden tot aan het einde van het project. Zonder deze vermindering van het debiet zal de verlaging verder oplopen. Kunnen we het verloop van het debiet gedurende de gehele duur van de bemaling berekenen?

De analytische formules die ons ter beschikking staan vergen alle een bepaald onttrekkingsdebiet, waarvan de verlaging dna het gevolg is. Er zijn analytische formules die de verlaging opleggen en daarbij het bemalingsdebiet uitrekenen. In een numeriek model is het juist eenvoudig om de gewenste verlaging op te leggen en de daarvoor benoigde onttrekking uit te rekenen. Met analytische formules zullen we het bewuste debiet iteratie moeten berekenen.

Het is wel eenvoudig om op basis van analytische formules de verlaging te berekenen bij een debiet dat willekeurig in de tijd verloopt. Dit kan met convolutie. Convolutie is een efficiÃ«nte manier van superpositie, waarbij het debiet in het verleden wordt gewogen met de impuls respons, de block respons of te staprespons.

De impuls-respons is het effect van een puls onttrekking, dat wil zeggen de onttrekking van een eenheid water P top t=0. De stap respons is de respons tegenvolge van een constante eenheids-sonttreking (Q = 1 m3/d)vanaf t=0. De meeste analytische tijdsafhankelijke oplossingen voor onttrekking zijn stap responsen wanneer je ze door de onttrekking deelt. De block respons is de response van een constante onttrekking gedurende de tijdstap waar men mee wil werken. Het is de staprespose vanaf t=0 verminderd met dezelfde stapresponse voot $t>dt$, de tijdstap

Voor een uniforme onttreking in een oneindig uitgestrekte laag is de stap response gelijk aan de oplossing volgens Theis

$$SR = \frac{1}{4 \pi kD} W_t(u) = \frac 1 {4 \pi kD}\intop_u^\infty \frac{e^{-y}}{y} dy,\,\,\,u=\frac{r^2 S}{4 kD t}$$

$$ IR = \frac{\partial}{\partial t} SR = \frac Q {4 \pi kD} \frac{e^{-u}}{t}$$

$$ BR = \frac Q {4 \pi kD} \left[ W_t\left({u_t}\right) - W_t\left({u_{t-dt}}\right)\right] $$

De berekening van de verlaging door een willekeurig verlopend debie geschiedt het gemakkeijkst met de functie lfilter uit de module`scipy.signal`.

$$s(r,t) =  \frac{1}{4 \pi kD} \sum_{i=0}^\infty Q(t - \tau_i) BR(\tau_i)=\frac{1}{4 \pi kD} \mbox{lfilter}(BR(\tau, 1, Q(t-\tau))$$

Het werken met de blokrespons is doorgaans het meest effectief. Convolutie kan ook met de impuls response, maar dan zijn er hele kleine tijdstappen nodig om voldoende nauwkeurigheed te bereiken; werken met de blokrespons is exact voor de situatie dat de onttrekking per tijdstap (per dag) constant is, maar wel van dag tot dag kan variÃ«ren.

```python
kD, S, Q0 = 650, 0.2, 1200
r = 20


tauEnd, dtau = 50., 1
t = tau = np.arange(0., tauEnd, dtau); tau[0] = 1e-6
Q = np.zeros_like(t) + Q0
u = r ** 2 * S / (4 * kD * tau)
BR = 1 / (4 * np.pi * kD) * (Wt(u) - np.hstack((0, Wt(u[:-1]))))

sTh = Q / (4 * np.pi * kD) * Wt(u)
sBR = lfilter(BR, 1, Q)

ax = newfig("Verl. vlgs. Theis ber. met convolutie, kD={:.0f} m2/d, S={:.3g}".format(kD, S), "tijd [d]", "verlaging [m]")

clrs = cycle('brgmck')
ax.plot(t, sTh, color=next(clrs), label='Theis')
ax.plot(t, sBR, '.', color=next(clrs), label='Blok-respons,dtau={:.3g} d'.format(dtau))

dtaus = [1, 0.3, 0.1, 0.03]
clr = next(clrs)
for dtau in dtaus:
    t = tau = np.arange(0., tauEnd, dtau); tau[0] = 1e-6
    u = r ** 2 * S / (4 * kD * tau)
    Q = np.zeros_like(t) + Q0
    IR = 1 / (4 * np.pi * kD) *  np.exp(-u) / tau * dtau 
    sIR = lfilter(IR, 1, Q)
    ax.plot(t, sIR, color=clr, label='Impuls-respons, dtau={:.3g} d'.format(dtau))

ax.legend()
plt.show()
```

Conclusie:

De uitkomst van de convolutie met de blokrespons valt samen met de directe toepassing van de formule van Theis. Voor de convolutie met de steprespons is dat pas het geval wanneer de staplengte klein genoeg is, in dit geval ca. 0.03 d. De afwijking met Theis bij tijdstappen van 1 dag is heel groot.

We kunnen nu, met convolutie met de blokrespons nauwkeurig de verlaging bij elk willekeurig verloop van de onttrekking uittrekenen.

Enkele voorbeelden

1. Willekeurig verloop van de onttrekking

1. Onttrekking neemt exponentiaal af in de tijd, waarbij de onttrekking halveeft over een tijdvak T

```python
kD, S, Q0 = 650, 0.2, 1200
r = 20

dtau = 1.0
tauEnd = 365
t = tau = np.arange(0, tauEnd, dtau); tau[0] = 1e-6 
T = tauEnd / np.log(2)
Q  = np.ones_like(t) * Q0
Q1 = Q0 * (np.random.rand(len(t)) + 0.5)

B = np.ones(int(tauEnd / 25))
B /= len(B)
Q1 = filtfilt(B, 1, Q1, method='gust', irlen=len(B))
Q2 = np.interp(t, np.array([0., t[-1]]),np.array([Q0, Q0 / 2]))
Q3 = Q0 * np.exp(- t / T)

u = r ** 2 * S  / (4 * kD * tau)
BR = 1 / (4 * np.pi * kD) * (Wt(u) - np.hstack((0, Wt(u[:-1]))))
sT = Q0 / (4 * np.pi * kD) * Wt(u)
s0 = lfilter(BR, 1.0, Q)
s1 = lfilter(BR, 1.0, Q1)
s2 = lfilter(BR, 1.0, Q2)
s3 = lfilter(BR, 1.0, Q3)

fig, axs = plt.subplots(2, 1, sharex=True, figsize=(8, 8))
axs[0].set_title("Verschillend verloop van de onttrekkingkD={:.0f} m2/d, S={:.3g}".format(kD, S))
axs[0].set_ylabel("Q [m3/d']")
axs[1].set_ylabel("Verlaging [m]")
axs[1].set_xlabel("tijd [d]")
axs[0].grid()
axs[1].grid()

clrs = cycle('brgmck')

axs[0].plot(t,  Q, color=next(clrs), label='Q0={:.0f} m3/d'.format(Q0))
axs[0].plot(t, Q1, color=next(clrs), label='Q1 (random)')
axs[0].plot(t, Q2, color=next(clrs), label='Q2 (halveert linear')
axs[0].plot(t, Q3, color=next(clrs), label='Q3 (halveert exponentieel')

clrs = cycle('brgmck')

axs[1].plot(t, sT, 'k.', label='Theis (Q0={} m3/d'.format(Q0))
axs[1].plot(t, s0, color=next(clrs), label='Q0 (Q0={:.0f} m3/d)'.format(Q0))
axs[1].plot(t, s1, color=next(clrs), label='Q1 (random)')
axs[1].plot(t, s2, color=next(clrs), label='Q2 (halveert linear')
axs[1].plot(t, s3, color=next(clrs), label='Q3 (halveert exponentieel')

axs[0].legend()
axs[1].legend()

```

Bemaling die de vereiste verlaging bereikt van $t_0$ tot $t_1$ en dan handhaafd tot $t_2$, waarna de bemaling gestaakt wordt.

```python
kD, S, Q0 = 650, 0.2, 1200
r = 20
dtau, tauEnd = 1.0, 365.

t0, t1, t2 = 0, 7, 97

t = tau = np.arange(0, tauEnd, dtau)[1:]

u = r ** 2 * S  / (4 * kD * tau)
BR = 1 / (4 * np.pi * kD) * (Wt(u) - np.hstack((0, Wt(u[:-1]))))

fig, axs = plt.subplots(2, 1, sharex=True, figsize=(8, 8))
axs[0].set_title("Verschillende snelheiden exp. daling onttrekking. kD={:.0f} m2/d, S={:.3g}".format(kD, S))
axs[0].set_ylabel("Q [m3/d']")
axs[0].set_ylim(0, Q0)
axs[0].set_xscale('log')
axs[1].set_ylabel("Verlaging [m]")
axs[1].set_xlabel("tijd [d]")
axs[0].grid()
axs[1].grid()

clrs = cycle('brgmck')

for Q1 in Q0 / np.array([1, 1.1, 1.2, 1.3, 1.4, 1.5]):

    Q = np.ones_like(t) * Q0
    if Q1 == Q0:
        Q[t > t1] = (t1 - t0) / (t[t > t1] - t0) * Q0
        DQ = np.zeros_like(t)
        DQ[t > t1] = Q0 * (t[t > t1] - t1) / (t[t > t1] - t0)
        axs[0].plot(t, DQ, label='DQ')
    else:
        Q[t > t1] = Q1
        Q[t > t2] = 0.

    s = lfilter(BR, 1.0, Q)

    clr = next(clrs)
    axs[0].plot(t, Q, color=clr, label='Q0={:.0f} m3/d, Q1={:.0f} m3/d'.format(Q0, Q1))
    axs[1].plot(t, s, color=clr, label='Q0={:.0f} m3/d, Q1={:.0f} m3/d'.format(Q0, Q1))

ylim = axs[1].get_ylim()
axs[1].set_ylim(0, ylim[1])

axs[0].legend()
axs[1].legend()
```

# Bennett, GD, Patten EP (1964) Groudnwater hydraulics. Constant-head pumping test of a muliaquifers well to determine characteristics of individual aquifers.USGS Water-Supply Paper 1536-G, Washington 1962.

$$G(\beta) = \frac {4 \beta}{\pi} \intop_{x=0}^{\infty} xe^{-\beta x^2} \left[\frac \pi 2 + \arctan \frac{K_0(x)}{J_0(x)}\right]dx$$

$$ Q(t) = 2 \pi kD\,H \, G\left(\frac{kDt}{r^2S }\right)$$


Ferris, JG, Knowles DB, Brrown RH and Stallman RW (1962) Theory of aqufier tests. US Geologial Suvey Water Supply Paper 1536-E

```python
def integrand(x, beta):
    return 4 * beta / np.pi * x * np.exp(-beta * x ** 2) * (np.pi / 2 + np.arctan(K0(x) / J0(x)))

points = jn_zeros(0, 50)
eps = 1e-20
plt.title("Show the intgrand.")

for beta in [-0.1]: # np.logspace(-1, 4, 6):
    R = 0    
    for i, (xp1, xp2) in enumerate(zip(points[:-1], points[1:])):
        dR, dE = quad(integrand, xp1 + eps, xp2 - eps, args=(beta,))
        if np.abs(dR) < 1e-20:
            break
        R += dR
        print('xp1 = {:8.3f}, xp2 = {:8.3f}, dR = {:10.6g}, R = {:10.6g}, dE = {:10.6g}'.format(xp1, xp2, dR, R, dE))
        x = np.linspace(xp1 + eps, xp2 - eps, 1000)
        plt.plot(x, integrand(x, beta))
        if i == 1:
            break
# plt.ylim(-0.1, 0.1)
plt.show()

```

```python
def G(beta):
    """Benntett and Patten (1962) function for constant head in well.
    
    Parameters
    ----------
    beta = kD t / r^2 S
    
    >>>G(0.004)
    9.41    
    """
    points = jn_zeros(0, 15)
    def kernel(x, beta):
        """Return the function to be integrated."""
        return 4 * beta / np.pi * x * np.exp(-beta * x ** 2) * (np.pi / 2 + np.arctan(K0(x) / J0(x)))
    def g(beta): # Integrate the argument
        return quad(kernel, 0., points[-1], args=(beta,), limit=100, points=points)[0]
    gh = np.frompyfunc(g, 1, 1) # Vectorize function g(beta) so we can use arrays as input.
    return np.asarray(gh(beta), dtype=float)

G(0.004)


x = np.logspace(-3,3 , 121)
x = np.linspace(0, 10, 1201)
for beta in np.logspace(-1, 0, 2):
    plt.plot(x, integrand(x, beta), label='Integrand, beta={:.4g}'.format(beta))

plt.grid()
#plt.xscale('log')
plt.legend()
plt.show()
```
